```python
%matplotlib
import math
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import figure
from ipywidgets import interact

@interact
def seno(c=1):
    x = np.linspace(0, 2*math.pi, 100)
    y = np.sin(x*c)
    
    plt.plot(x, y)
    plt.ylabel('seno(x)')
    plt.xlabel('x')
    plt.title('Onda sinusoidal')
    
seno(c=2)
```
